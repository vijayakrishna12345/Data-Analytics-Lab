```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
```

```
# Boston Dataset
df = datasets.load_boston()
X = df.data
print(df)
```

```
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
#kmeans.labels_
kmeans.fit(X)
kmeans.predict(X)
labels = kmeans.labels_
print(labels)
```

```
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
plt.title("K Means", fontsize=14)
```

```
#Complete Linkage
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering().fit(X)
AgglomerativeClustering(affinity='euclidean',linkage='complete',n_clusters=2,pooling_func='deprecated')
print(clustering.labels_)
```

```
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
plt.title("Complete Linkage", fontsize=14)
```

```
#Single Linkage
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering().fit(X)
AgglomerativeClustering(affinity='euclidean',linkage='single',n_clusters=2,pooling_func='deprecated')
print(clustering.labels_)
```

```
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
plt.title("Single Linkage", fontsize=14)
```

```
#Average Linkage
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering().fit(X)
AgglomerativeClustering(affinity='euclidean',linkage='average',n_clusters=2,pooling_func='deprecated')
print(clustering.labels_)
```

```
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
plt.title("Average Linkagee", fontsize=14)
```

```
#Wards Linkage
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering().fit(X)
```

```
AgglomerativeClustering(affinity='euclidean', compute_full_tree='auto',
                        connectivity=None, distance_threshold=None,
                        linkage='ward', memory=None, n_clusters=2,
                        pooling_func='deprecated')
print(clustering.labels_)
```

```
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
plt.title("Wards Linkage", fontsize=14)
```
